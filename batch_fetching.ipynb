```python
import time
try:
    from googlesearch import search
except ImportError:
    print("No module named 'google' found")

# Set the desired number of search results per batch
results_per_batch = 20

# Set the total number of search results you want to fetch
total_results = 100

# Define your search query
query = "ibm"

# Perform the search and get the results
results = []

iterations = total_results // results_per_batch  # Number of iterations needed

for i in range(iterations):
    batch_start = i * results_per_batch

    # Fetch results for the current batch
    batch_results = list(search(query, tld="co.in", num=results_per_batch, start=batch_start, pause=5))

    # Append batch results to the main results list
    results.extend(batch_results)

    # Print the batch results
    for j, result in enumerate(batch_results, start=batch_start+1):
        print(f"Result {j}: {result}")

    # Pause between batches
    if i < iterations - 1:
        time.sleep(10)  # Adjust the pause duration as needed

# Print the total number of results fetched
print(f"Total results fetched: {len(results)}")
```

```python
import time
from googlesearch import search

query = "ibm"
num_results = 20
batch_size = 5
pause_between_batches = 5  # Pause in seconds between batches (adjust if needed)

results = []

# Perform the search in batches
for i in range(0, num_results, batch_size):
    # Set the number of results
    num = min(batch_size, num_results - i)

    # Perform the search and store the results
    try:
        search_results = search(query, num=num, pause=pause_between_batches)
        results.extend(search_results)
    except Exception as e:
        print(f"An error occurred: {e}")
    
    # Delay between batches to avoid "Too Many Requests" error
    time.sleep(pause_between_batches)

# Print the search results
for result in results:
    print(result)
```
